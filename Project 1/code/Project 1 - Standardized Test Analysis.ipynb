## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. These scores are used along with other materials such as grade point average (GPA) and essay responses to determine whether a potential student will be accepted to university.

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from students' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

- The SAT has two sections of the test:   
    1. Evidence-Based Reading and Writing  
    2. Math  
[*Source*](https://www.princetonreview.com/college/sat-sections)  
- The ACT has 4 sections:  
    1. English  
    2. Math
    3. Reading
    4. Science  
    [*Source*](https://www.princetonreview.com/college/act-sections#:~:text=The%20ACT%20includes%20the%20following,you%20register%20for%20the%20ACT.)  
    
They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

```python
# Code:

def mean_calc(values):
    return sum(values) / len(values)
    
test = [2,4,6,8,10]
mean_calc(test)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:

def std_dev_calc(values):
# finding mean value
    mean = mean_calc(values)
# finding sum squares of difference of mean and each value
    diff_sqrd = 0
    for i in range(len(values)):
        diff_sqrd += (values[i] - mean) ** 2
# finding standard deviation
    std_dev = (1 / len(values) * diff_sqrd ) ** 0.5
    return std_dev

test = [2,4,6,8,10]
std_dev_calc(test)
```

Comparing our standard deviation calculations, there seems to be a difference between standard deviation calculated by the describe() function and the one we obtained using our std_dev_calc() function.

The reason for such a difference lies in the way the standard deviation formula is derived for each of the two methods.

The forumla provided for out standard deviation calculator is known as the [Population Standard Deviation](https://www.calculator.net/standard-deviation-calculator.html).  

The formula used by the describe() function however, uses the formula known as the [Sample Standard Deviation](https://www.thoughtco.com/population-vs-sample-standard-deviations-3126372).

Which one should we use? Datapoints in our datasets used serve to inform us about the nation's exam takers. By definition, it includes the entire population of exam takers. Therefore, a population standard deviation would be used.

We may read more about the use cases for the two different standard deviations [here](https://statistics.laerd.com/statistical-guides/measures-of-spread-standard-deviation.php).

### Regarding Dataset limitations

From out datasets, each datapoint represents a sample from a population.

Using the ACT Math Scores for example:
Population: The scores of all test-takers nation-wide.
Population mean: National average of the Math test (Derived from dividing total scores with total number of test takers)
Sample mean: Average of the math test for an individual state (out of 51)

Based on the Central Limit Theorem, we generally assume data sampled from a population to be normally distribution. This does not seem the case based on our frequency distributions shown above.

Central Limit Theorem does not hold in our project due to the following reasons:
1. We do not know the distribution of the underlying data for each state. Although strongly skewed distributions requires larger sample sizes upwards of 30, it could mean that having 51 samples (from all states) might still be insufficient for Central Limit Theorem to take effect if the underlying data is not normal.

2. For Central Limit Theorem to hold, samples taken from the population must be random. Samples in our case are randomised as underlying factors for an individual state would heavily affect test scores and participation. (i.e Demographic, state policies, quality of school, etc)

For all test scores (total and section), ACT distributions tend to lean closer towards normal distribution compared to SAT.
Possible reasons for such an observation could be due to larger number of states mandating participation for the ACT compared to the SAT. As only scores for test takers are captured in the distributions, ACT would have less selection bias than the SAT. Selection bias in this case refers to the self-selection bias, in which test takers who are 'better prepared' and 'more intelligent' would choose to take the exams, skewing the underlying result.

To conclude, the Central Limit Theorem does not hold as we do not have randomised samples, which is a key assumption for the Central Limit Theorem. This is true for distributions regarding Math, Reading and participation rates as none of them seem normal with significant skew and kurtosis. As the number of test takers varies between states, our sample size is also varied which further limit the Central limit Theorem assumption.

Despite the limitations presented in our datasets, we can still improve the results of statistical testing (although still largely inaccurate) by dropping states that do not have mandatory testing (less than 95% participation). This would improve how much our samples are randomised to a degree, however it would still be insufficient to ensure the Central Limit Theorem hold.

The optimal way to conduct statistical inference on this subject matter is to obtain datasets representative of all college students, coupled with their decision to participate in which of the test and their scores in one or both of the tests. That way we are able to create randomised samples with equal number of students in each sample.



### Can we find out which is the easier test to take for a potential test taker from our datasets?

We will not be able to statistically conclude if the ACT was easier than the SAT or vice versa. To do so we have to compare the means of both ACT and SAT scores. The inference will only hold if their distributions were normal which was not the case with our datasets.
